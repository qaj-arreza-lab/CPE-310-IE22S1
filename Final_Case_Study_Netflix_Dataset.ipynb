⋆⭒˚🍓｡⋆𝓓𝓾𝓻𝓪𝓽𝓲𝓸𝓷 & 𝓢𝓮𝓪𝓼𝓸𝓷𝓼⋆⭒˚🍓｡⋆

Parse duration into numeric minutes (for movies) and number of seasons (for shows). What is the average movie length? The average number of seasons per TV show?

```python
# Duration & Seasons
nflx = pd.read_csv('/content/netflix_titles.csv')
```

```python
nflx['duration_num'] = nflx['duration'].str.extract(r'(\d+)').astype(float)

movies = nflx[nflx['type'] == 'Movie']
tv_shows = nflx[nflx['type'] == 'TV Show']

avg_movie_length = movies['duration_num'].mean()
avg_seasons = tv_shows['duration_num'].mean()

print("Average movie length (minutes):", avg_movie_length)
print("Average number of seasons:", avg_seasons)
```

⋆⭒˚🍓｡⋆𝓖𝓮𝓷𝓻𝓮 𝓐𝓷𝓪𝓵𝔂𝓼𝓲𝓼⋆⭒˚🍓｡⋆

Which genre has the highest average release year? What might that indicate?

```python
genres = nflx.assign(listed_in=nflx['listed_in'].str.split(', ')).explode('listed_in')

genre_avg_year = (
    genres.groupby('listed_in')['release_year']
    .mean()
    .sort_values(ascending=False)
)

top_genre = genre_avg_year.idxmax()
top_value = genre_avg_year.max()

print(top_genre, top_value)
```

⋆⭒˚🍓｡⋆𝓣𝓮𝓶𝓹𝓸𝓻𝓪𝓵 𝓣𝓻𝓮𝓷𝓭𝓼⋆⭒˚🍓｡⋆

Are there spikes or drops? Compare release_year vs. year_added: how long after release does Netflix typically acquire content?

```python
import matplotlib.pyplot as plt

nflx['date_added'] = pd.to_datetime(nflx['date_added'].astype(str).str.strip(), errors='coerce')
nflx['year_added'] = nflx['date_added'].dt.year

trend_comparison = nflx[['release_year', 'year_added']].apply(pd.Series.value_counts).sort_index()
trend_comparison.loc[2010:2021].plot(kind='line', figsize=(10, 5))
plt.title('Content Release vs. Netflix Addition (2010-2021)')
plt.ylabel('Count of Titles')
plt.show()

nflx['acquisition_gap'] = nflx['year_added'] - nflx['release_year']

gap_analysis = nflx[nflx['acquisition_gap'] >= 0]['acquisition_gap'].value_counts().sort_index().head(10)
gap_analysis.plot(kind='bar', color='orange')
plt.title('Typical Time Gap Before Netflix Acquisition')
plt.xlabel('Years After Release')
plt.ylabel('Number of Titles')
plt.show()

print(f"Median Acquisition Gap: {nflx['acquisition_gap'].median()} years")
```

⋆⭒˚🍓｡⋆𝓐𝓹𝓹𝓵𝔂𝓲𝓷𝓰 𝓕𝓾𝓷𝓬𝓽𝓲𝓸𝓷𝓼⋆⭒˚🍓｡⋆

Write a function that, given a director’s name, returns their titles sorted by release_year. Create a reusable function to plot the top N categories for any categorical column.

```python
def print(netflix, director_name):
  result = netflix[netflix["director"] == director_name][["title", "release_year"]]
  return result.sort_values(by = "release_year" )
```

```python
import matplotlib.pyplot as plt

def plot_top_categories(nflx, column, n=10):
    top_categories = nflx[column].value_counts().head(n)

    top_categories.plot(kind='bar')
    plt.title(f"Top {n} Categories in {column}")
    plt.xlabel(column)
    plt.ylabel("Count")
    plt.show()
```

```python
plot_top_categories(nflx, "country", 5)
```
